```python
import json
from pprint import pprint
```

```python
with open('../datasets/generated/organization_en_final.json', encoding='utf-8') as fp:
    organizations = json.load(fp)
```

```python
unit_types = set()
sub_types = set()
for organization in organizations['value']['data']:
    unit_types.add(organization['unitType'])
    sub_types.add(organization['subType'])
```

```python
unit_types
```

```python
sub_types
```

```python
unit_division = {}
for unit in unit_types:
    unit_division[unit] = []
for organization in organizations['value']['data']:
    unit_division[organization['unitType']].append(organization)
    del unit_division[organization['unitType']][-1]['unitType']
pprint(unit_division)
```

```python
for key in [*unit_division]:
    print(key, len(unit_division[key]))
```
